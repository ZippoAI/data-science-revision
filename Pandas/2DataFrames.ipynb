```python
import numpy as np
import pandas as pd
```

## Creating a data frame

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 41],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}

df = pd.DataFrame(data)

df
```

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]

labels = ['Name', 'Age', 'City', 'Salary']
df2 = pd.DataFrame(data_list, columns=labels)

df2
```

## Selection and indexing of columns

```python
## Selecting columns

df2['Name']
```

```python
## Selecting two columns

df2[['Name', 'Salary']]
```

## Creating new columns


```python
df2['Designation'] = ['Doctor', 'Engineer', 'Teacher', 'Pilot']

df2
```

## How to remove a column


```python
df2.drop('Designation', axis=1, inplace=True) # use inplace true to save it

df2
```

```python
df3 = df2

df3
```

```python
df3 = df3.drop('Name', axis=1)

df3
```
